```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Content_df = pd.read_csv(r'C:\Users\USER\Documents\Data_Analytics_Porfolio\Accenture Virtual Internship\Contentpy.csv')
```

```python
Content_df= Content_df.drop(['User ID','URL'],axis = 1)
```

```python
Content_df.rename(columns={'Type':'ContentType'},inplace = True)
Content_df
```

```python
Reactions_df = pd.read_csv(r'C:\Users\USER\Documents\Data_Analytics_Porfolio\Accenture Virtual Internship\Reactionspy.csv')
```

```python
Reactions_df= Reactions_df.drop(['User ID'],axis = 1 )
Reactions_df.rename(columns={'Type':'ReactionType'},inplace = True)
Reactions_df
```

```python
Reactions_df.isnull().sum()
```

```python
ReactionTypes_df = pd.read_csv(r'C:\Users\USER\Documents\Data_Analytics_Porfolio\Accenture Virtual Internship\ReactionTypespy.csv')
```

```python
ReactionTypes_df.rename(columns={'Type':'ReactionType'},inplace = True)
ReactionTypes_df
```

```python
df = pd.merge(Reactions_df,Content_df, how= 'left',on= 'Content ID')
df = pd.merge(df,ReactionTypes_df, how ='left', on='ReactionType')
df
```

```python
df = df.drop(['Unnamed: 0_x','Unnamed: 0_y','Unnamed: 0'], axis=1)
df
```

```python
df['Category'] = df['Category'].str.strip().str.replace('"', '')

corrections = {
    'animals': 'Animals',
    'healthy eating': 'Healthy Eating',
    'public speaking': 'Public Speaking',
    'veganism': 'Veganism'
}

df['Category'] = df['Category'].replace(corrections)

df['Category'] = df['Category'].str.title()
```

```python
result = df.groupby('Category')['Score'].sum().sort_values(ascending = False).reset_index()
# result
Top_5 = result.head()
Top_5
```

```python
# df.to_csv('CleanedDataset.csv',index=False)
```

```python
sns.barplot(x = 'Category', y = 'Score' , data = Top_5, palette = ['#4682B4'])

plt.title('Top 5 categories by popularity Score')
plt.xlabel('categories ')
plt.ylabel('popularity Score')
plt.xticks(rotation = 90)
plt.show()
```

```python
Category = Top_5['Category']
Score = Top_5['Score']
colours = ['#4682B4']
explode = (0.05, 0.05, 0.05, 0.05, 0.05)

plt.figure(figsize=(10, 6)) 
_, texts, autotexts = plt.pie(Score,
                      labels=Category,
                      colors = colours,         
                      autopct='%1.1f%%', 
                      pctdistance=0.85,
                      explode=explode,
                      startangle=60)

for autotext in autotexts:
    autotext.set (color='white', fontsize=12, fontweight='bold')
[text.set(color='black', fontsize=13, fontweight='bold') for text in texts]
    
plt.title('Popularity % share from top 5 Category',fontsize=15, fontweight='bold')

plt.show()
```

```python
unqiue_category = df['Category'].value_counts().reset_index()
unqiue_category.columns = ['Category','Score']
unqiue_category
```

```python
sns.barplot(x = 'Category', y = 'Score' , data = unqiue_category, palette = ['#4682B4'])

plt.title('Unique Categories')
plt.ylabel('Value Count')
plt.xlabel('Unique Categories')
plt.xticks(rotation = 90)

plt.show()
```
